```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
#from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import confusion_matrix
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Embedding
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
spam = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/NLP/6.ML e DL na Prática/spam.csv')
spam.head()
```

#Tratamento

```python
labelencoder = LabelEncoder()
y = labelencoder.fit_transform(spam['Category']) # transforma as classes em valores númericos
print(y)
```

```python
 mensagens = spam['Message'].values
 X_treino, X_teste, y_treino, y_teste = train_test_split(mensagens, y, test_size=0.3)
```

#Tokenização

```python
token = Tokenizer(num_words=1000)
token.fit_on_texts(X_treino)
```

```python
X_treino = token.texts_to_sequences(X_treino)
X_teste = token.texts_to_sequences(X_teste)
```

```python
print(len(X_treino))
print(len(X_teste))
```

```python
print(X_treino)
```

```python
X_treino = pad_sequences(X_treino, padding="post", maxlen=500)# definindo o tamanho da matrix
X_teste = pad_sequences(X_teste, padding="post", maxlen=500)
```

```python
print(X_treino)
```

```python
modelo = Sequential()
modelo.add(Dense(units=10, activation='relu', input_dim=X_treino.shape[1])) #camada de entrada
modelo.add(Dropout(0.1)) # reduz as sinapses para evitar overfitting
modelo.add(Dense(units=8, activation='relu')) #camada oculta
modelo.add(Dropout(0.1)) # reduz as sinapses para evitar overfitting
modelo.add(Dense(units=1, activation='sigmoid')) #camada de saída
```

#Compilação

```python
#compilação é a montagem da estrutura
modelo.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy']) #adam = stochastic gradient descent
modelo.summary()
```

#Treino

```python
#batch_size = a cada quantos registros vai atualizar os dados da rede
modelo.fit(X_treino, y_treino, epochs=20, batch_size=10, verbose=True, validation_data=(X_teste, y_teste)) #esse modelo faz autoavaliação
```

#Avaliando performance

```python
loss, accuracy = modelo.evaluate(X_teste, y_teste)#avaliando perfomance com dados de teste
print('Loss: ', loss)
print('Accuracy: ', accuracy)
```

```python
nova_previsao = modelo.predict(X_teste)
print(nova_previsao) #a ativação da camada usa sigmoid, por tanto a previsão terá um valor númerico entre 0 e 1
```

```python
prev = (nova_previsao > 0.5)# dessa forma o resultado da previsão será True ou False
print(prev)
```

```python
cm = confusion_matrix(y_teste, prev)
print(cm)
```
